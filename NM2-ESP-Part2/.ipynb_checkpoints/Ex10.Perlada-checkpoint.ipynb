```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import linregress
```

```python
def compute_mean(data):
    """
    Calculate the mean of the dataset.
    """
    total = 0
    for d in data:
        total += d
    mean = total / len(data)
    return mean

def compute_stdev(data, mean):
    """
    Calculate the standard deviation of the dataset.
    """
    squared_diff = 0
    for d in data:
        squared_diff += (d - mean) ** 2
    variance = squared_diff / (len(data)-1)
    std_dev = variance ** 0.5
    return std_dev

def plot_tseries(data, mean, stdev):
    """
    Plot the time series data with a regression line and mean value.
    """
    time = np.arange(len(data))  # Assuming time points are just indices
    slope, intercept, r_value, p_value, std_err = stats.linregress(time, data)
    regression_line = slope * time + intercept
    
    plt.figure(figsize=(10, 6))
    plt.plot(time, data, color='black', linestyle='-', label='ERA5')
    plt.plot(time, regression_line, color='red', linestyle='--', label='Regression Line')
    plt.axhline(y=mean, color='grey', linestyle='--', linewidth=2, label='Mean')  # Plot mean line
    plt.title('2m air Temperature @ Trieste 1990-2023')
    plt.xlabel('Time')
    plt.ylabel('Temperature (K)')
    plt.legend()
    plt.grid(True)

def compute_sample_slope(x, y):
    """
    Compute the slope of the linear regression line.
    """
    n = len(x)
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    numerator = np.sum((x - mean_x) * (y - mean_y))
    denominator = np.sum((x - mean_x)**2)
    slope = numerator / denominator
    return slope

def compute_standard_deviation_error(x, y, slop):
    """
    Compute the standard error of the slope.
    """
    n = len(x)
    
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    
    # Compute the numerator of the variance term
    numerator = np.sum((x - mean_x) ** 2 * (y - mean_y) ** 2)
    
    # Compute the denominator
    denominator = np.sum((x - mean_x) ** 2)
    
    # Compute the variance
    var = np.sum((y - mean_y) ** 2) - (numerator / denominator)
    
    # Compute the variance of the error
    var_err = var / (n - 2)
    
    # Compute the standard error of the slope
    sb_1 = np.sqrt(var_err / denominator)
    
    return sb_1

def compute_t_statistic(slope, sb_1):
    """
    Compute the t-statistic.
    """
    t_statistic = slope / sb_1
    return t_statistic

def find_critical_t_value(df, alpha=0.05):
    """
    Find the critical t-value from the t-distribution table.
    """
    critical_value = stats.t.ppf(1 - alpha/2, df)
    return critical_value

def compute_annual(data):
    """
    Compute annual means from monthly means.
    """
    # Reshape the data to 12 months per year
    data_reshaped = data.reshape(-1, 12)
    # Compute the mean along the months axis
    annual_means = np.mean(data_reshaped, axis=1)
    return annual_means
```

```python
# Load txt file and call functions
data_in = np.loadtxt('ERA5_2m_SAT_TS_1990_2023.txt', usecols=2)
time = np.arange(len(data_in))  # Assuming time points are just indices

mean = compute_mean(data_in)
stdev = compute_stdev(data_in, mean)

# Plot data with regression line and mean line
plot_tseries(data_in, mean, stdev)

plt.show()
```

```python
# Compute sample slope and standard deviation of the errors
slope = compute_sample_slope(time, data_in)
sb_1 = compute_standard_deviation_error(time, data_in, slope)

# Compute t-statistic
t_statistic = compute_t_statistic(slope, sb_1)

# Degrees of freedom
# df = len(data_in) - 1
df = 100
```

```python
# Calculate slope and standard error using stats.linregress
b1_linregress, intercept_linregress, r_value_linregress, p_value_linregress, std_err_linregress = linregress(time, data_in)
sb1_linregress = std_err_linregress

# Print the slope and standard error from our custom functions and stats.linregress
print("Using custom functions:")
print("Slope (b1):", slope)
print("Standard Error (sb1):", sb_1)

print("\nUsing scipy functions:")
print("Slope (b1):", b1_linregress)
print("Standard Error (sb1):", sb1_linregress)

# Compare t-score and p-value from our custom test with stats.linregress
print("\nUsing custom functions:")
print("t-score:", t_statistic)
print("p-value:", stats.t.sf(np.abs(t_statistic), df) * 2)  # two-tailed test

print("\nUsing scipy functions:")
print("t-score:", b1_linregress / sb1_linregress)
print("p-value:", p_value_linregress)
```

```python
# Find critical t-value
critical_t_value = find_critical_t_value(df)

# Print the calculated t-statistic and critical t-value
print("t-statistic:", t_statistic)
print("Critical t-value:", critical_t_value)

print("\n")

# Determine significance
if abs(t_statistic) > critical_t_value:
    print("The result of the test is statistically significant.")
else:
    print("The result of the test is not statistically significant.")
```

To determine whether the increase in surface air temperature in Trieste from 1990 is statistically significant, we need to consider the significance of the slope of the regression line fitted to the monthly time-series. If the absolute value of the t-statistic is greater than the critical t-value (for a significance level of 0.05), then we can conclude that the slope is statistically significant.

Based on the t-test we performed, the absolute value of the t-statistic (1.524509463467757) is less than the critical t-value (1.9839715184496334) (for a significance level of 0.05), thus we can conclude that the slope is not statistically significant. This indicates that the increase in surface air temperature is not statistically significant in the monthly time-series.

The lack of statistical significance in the monthly time series may be due to the inherent noise and variability present in the data. This noise can obscure any underlying trend, making it more challenging to detect statistical significance. Additionally, monthly data may exhibit strong seasonal patterns. These seasonal fluctuations can contribute to the overall variability in the data and make it harder to discern long-term trends.

### Annual Mean 

```python
# Compute the annual means
annual_means = compute_annual(data_in)

# Plot the annual mean time-series and the regression line
years = np.arange(1990, 2024)  # Assuming starting year is 1990
slope_annual, intercept_annual, r_value_annual, p_value_annual, std_err_annual = stats.linregress(years, annual_means)
regression_line_annual = slope_annual * years + intercept_annual

plt.figure(figsize=(10, 6))
plt.plot(years, annual_means, color='blue', label='Annual Mean Time Series')
plt.plot(years, regression_line_annual, color='red', linestyle='--', label='Regression Line')
plt.title('Annual Mean Time Series of 2m Air Temperature with Regression Line')
plt.xlabel('Year')
plt.ylabel('Temperature (K)')
plt.legend()
plt.grid(True)
plt.show()

# Step 3: Test the slope for significance
df_annual = len(years) - 1

# Compute sample slope and standard deviation of the errors
slope = compute_sample_slope(years, annual_means)
sb_1 = compute_standard_deviation_error(years, annual_means, slope)
# Compute t-statistic
t_statistic = compute_t_statistic(slope, sb_1)

print("Using custom functions:")
print("Slope (b1):", slope)
print("Standard Error (sb1):", sb_1)
print("t-score:", t_statistic)
print("p-value:", stats.t.sf(np.abs(t_statistic), df) * 2)  # two-tailed test


# t_statistic_annual = slope_annual / (std_err_annual / np.sqrt(len(years)))
t_statistic_annual = slope_annual / std_err_annual
critical_t_value_annual = find_critical_t_value(df_annual)

print("\nUsing scipy functions:")
print("Slope (b1):", slope_annual)
print("Standard Error (sb1):", std_err_annual)
print("t-statistic (Annual):", t_statistic_annual)
print("Critical t-value (Annual):", critical_t_value_annual)
print("p-value:", p_value_annual)

print('\n')

# Determine significance
if np.abs(t_statistic_annual) > critical_t_value_annual:
    print("The slope is significantly different from zero.")
    print("The slope of the test is statistically significant.")
else:
    print("The slope is not significantly different from zero.")
    print("The slope of the test is not statistically significant.")
```

To determine whether the increase in surface air temperature in Trieste from 1990 is statistically significant, we need to consider the significance of the slope of the regression line fitted to the annual mean time-series. If the absolute value of the t-statistic is greater than the critical t-value (for a significance level of 0.05), then we can conclude that the slope is statistically significant.

Based on the t-test we performed, the absolute value of the t-statistic (5.867072328716281) is greater than the critical t-value (2.034515297449338) (for a significance level of 0.05), thus we can conclude that the slope is statistically significant. This indicates that the increase in surface air temperature is statistically significant, suggesting either an increase or a decrease over time.

Aggregating temperature data into annual means may reduce noise and enhance the detectability of underlying trends compared to monthly data, which often exhibit higher levels of variability and noise. Annual means provide greater statistical power to detect significant trends compared to the monthly time series. Moreover, the annual aggregation may help mitigate the impact of seasonal fluctuations present in monthly data, making it easier to discern long-term patterns.
