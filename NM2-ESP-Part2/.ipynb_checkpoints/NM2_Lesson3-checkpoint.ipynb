```python
import iris
import matplotlib.pyplot as plt

from ds_module import *
```

##### Loading the two maps. 

```python
#Load high resolution dataset (ERA5 reanalysis)
D1=iris.load_cube('data/MSLP_ERA5_1950_2014_mean_30km.nc')
print(D1)
```

```python
#Load lower resolution dataset (UKESM model outputs)
D2=iris.load_cube('data/MSLP_UKESM_1950_2014_mean_100km.nc')
print (D2)
```

##### Looking on the resolution of the two maps, we can see that they have different resolution and performing operation will be problematic. **Solution:** re-gridding!

```python
print (D1.coords('latitude'))
print (D2.coords('latitude'))
```

##### Visualization of the two maps. We can visible see the their difference in resolution.

```python
create_map(D1, title ='Mean Sea Level Pressure')
```

```python
create_map(D2, title ='Mean Sea Level Pressure')
```

##### Performing some re-gridding. In this case, we want to make the coarser map into more high resolution map (Linear)

```python
D2_hireso = regrid(D2, D1, binary=True)
```

##### Let us have a look at the latitude coordinate. Now, we can see that both maps have the same resolution after re-gridding. 

Map 1 resolution details:

```python
print (D1.coords('latitude'))
```

Map 2 re-gridded and we can see the updated resolution details:

```python
print (D2_hireso.coords('latitude'))
```

Visualizing the new re-gridded map.

```python
print('Re-gridded map')
create_map(D2_hireso, title ='Mean Sea Level Pressure')
```

```python
anomaly_map = D1 - D2_hireso
```

```python
print('Anomaly map')
create_map(anomaly_map, title ='Mean Sea Level Pressure Anomaly')
```

##### Extracting the grid-points values of interest location using interpolation 

```python
lat = [41.9, 37.8, -16.5, -33.86]
lon = [12.5, 122.4, 68.12, 151.20]
```

##### We tried both both scheme in interpolation and obtain:

```python
dat1 = interpolationM(anomaly_map, lat, lon)
dat1
```

##### In this case, we want to use the other re-gridding scheme, the Nearest Neigbhor scheme.

```python
D2_hireso2 = regrid(D2, D1, binary=False)
```

```python
print('Re-gridded map')
create_map(D2_hireso2, title ='Mean Sea Level Pressure')
```

```python
anomaly_map2 = D1 - D2_hireso2
```

```python
print('Anomaly map')
create_map(anomaly_map2, title ='Mean Sea Level Pressure Anomaly')
```

#### Again, we use both interpolation scheme for finding the grid-values. 

```python
dat2 = interpolationM(anomaly_map2, lat, lon)
dat2
```

##### From values alone, it is difficult to determine the difference of both schemes, maybe better if we viewed both anomaly maps with different scheme side by side.

```python
create_double_map(anomaly_map, anomaly_map2, title1 = 'Linear', title2 = 'Nearest Neighbor')
```

##### From the above plots, we can clearly see the difference of the two scheme. For Linear scheme, we can see that the transitions of values is very smooth (see the colors in the Pacific oceans), while the Nearest Neighbhor scheme we can see a jump of color and/or single color of patch (sea the sea below Australia).

###### Additionally, we can also look the difference formed with the re-gridding of both schemes, 

```python
create_double_map(D2_hireso, D2_hireso2, title1 = 'Linear', title2 = 'Nearest Neighbor')
```

##### With the above plot, its kinda difficult to distinguished the difference. 

##### But in general, we now see the difference of the two schemes and it will boils down in the end on what do you want to achieve. Both method have its own, advantage and disadvantages.
