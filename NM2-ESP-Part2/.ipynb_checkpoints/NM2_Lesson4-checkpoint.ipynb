```python
import iris
import matplotlib.pyplot as plt
```

```python
from ds_module import *
```

```python
#Load Netcdf dataset
SST=iris.load_cube('data/PI_CMIP6_HadGEM3_100yrs.nc')
print(SST)
```

```python
#Extract Ocean Basin
where='NP' #NA= North Atlantic, NP= North Pacific , SO= Southern Ocean
SST=extract_area(SST, where=where)
```

```python
#Compute area-weighted annual time series
SST = area_weighted(SST)
print(SST)
```

```python
#Plot SST time-series 
plot_tseriesv2(SST.data, label='PI Control Run', color='black', linestyle='-', xlabel='Years', ylabel='SST (degC)', title='100 years of SST variability, {where}'.format(where=where), date=False)
```

```python
#Detrend the time-series 
SST=my_function_to_detrend(SST)
```

```python
#Plot SST time-series 
plot_tseriesv2(SST.data, label='PI Control Run', color='black', linestyle='-', xlabel='Years', ylabel='SST (degC)', title='100 years of SST variability (detrended), {where}'.format(where=where), date=False)
```

```python
#Call functions to compute 1D Power Spectrum (PWS)
S, nk=my_function_to_compute_FFT_1D(SST.data) 
```

```python
#Function to plot PWS. It takes as arguments the PWS (S), the frequencies (nk), the plot title, 
#and a logical switch: period=False for frequencies on x-axis, period=True for period on x-axis. 
Plot_S_1D(S, nk, title='PWS {where}'.format(where=where), period=True)

plt.show()
```
