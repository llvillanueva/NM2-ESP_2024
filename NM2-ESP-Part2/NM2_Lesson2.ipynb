```python
import numpy as np
import matplotlib.pyplot as plt

from ds_module import *
```

```python
############## Load datasets #############
#Load txt file and call functions
data_in=np.loadtxt('data/ERA5_2m_SAT_TS_1990_2023.txt', usecols=2)
```

```python
mean=compute_mean(data_in)
stdev=compute_std(data_in, mean)
```

```python
#Plot data with regression line and print out 'slope, std_err, p_value'

plot_tseries(data_in, mean, stdev, label='ERA5',
             linestyle='-', color='red', title='2m air Temperature @ Trieste 1990-2023')
```

```python
slope, std_err, t_score, p_value, t_crit = t_student(data_in)
```

```python
# Determine significance
if abs(t_score) > t_crit:
    print("The result of the test is statistically significant.")
else:
    print("The result of the test is not statistically significant.")
```

### Annual Mean

```python
#Plot data with regression line and print out 'slope, std_err, p_value'

plot_tseries(annual_means, mean, stdev, label='ERA5',
             linestyle='-', color='red', title='Annual means of 2m air Temperature @ Trieste 1990-2023')
```
