### Exercise on CTCS RAW scheme
- Write a program to integrate the linear advection equation in
\begin{equation}
    \frac{\partial \phi}{\partial t} + u \frac{\partial \phi}{\partial x} = 0
\end{equation}
using the CTCS scheme in the domain $0 \leq x \leq 1000\,m$ with advection velocity $u = 0.475\,m\,s^{-1}$ . Let $\Delta x = 0.125\,m$ and assume periodic boundary conditions. Assume the initial shape to be:
$$ \phi(x, 0)=  \left\{
\begin{array}{ll}
      0.0 \quad \mathrm{for} \quad x<400 \\
      1.00 \quad \mathrm{for} \quad 400 \leq x < 500 \\
      2.00 \quad \mathrm{for} \quad 500 \leq x < 600 \\
      0.0 \quad \mathrm{for} \quad x > 600 \\
\end{array} 
\right. $$ 

- Integrate forward and show solutions from $t=0\,s$ to $t = 2000\,s$ every $500\,s$ and explain the characteristics of the solution.
- Repeat the exercise implementing the RA filter with $\alpha = 0.10$.
- Repeat the exercise implementing the RAW filter with $\alpha = 0.05$, $\beta = 0.53$.

**Hint: Remember the CFL condition!**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class AdvectionSimulation:
    def __init__(self, x0, x1, t0, t1, u, dx, dt, tp, alpha, beta):
        self.x0 = x0  
        self.x1 = x1 
        self.t0 = t0
        self.t1 = t1
        self.u = u
        self.dx = dx
        self.dt = dt
        self.tp = tp
        self.alpha = alpha
        self.beta = beta
        self.c = u * dt / dx
        self.nx = round((x1 - x0) / dx) + 1
        self.x = np.linspace(x0, x1, self.nx)
        self.phi = np.array([self.phi0(xi) for xi in self.x])
        self.phi_old = np.copy(self.phi)
        self.phi_now = np.copy(self.phi)
        self.d = np.copy(self.phi)

    def phi0(self, x):
        if x < 400.0 or x > 600.0:
            return 0.0
        elif x >= 400 and x < 500:
            return 1.0
        else:
            return 2.0

    def ftfs(self, phi_now):
        return (1 + self.c) * phi_now - self.c * np.roll(phi_now, -1)

    def ftbs(self, phi_now):
        return (1 - self.c) * phi_now + self.c * np.roll(phi_now, 1)

    def ctcs(self, phi_old, phi_now):
        return phi_old - self.c * (np.roll(phi_now, -1) - np.roll(phi_now, 1))

    def plot(self):
        t = self.t0
        if self.u > 0:
            self.phi_now = self.ftbs(self.phi_old)
        else:
            self.phi_now = self.ftfs(self.phi_old)            
        plt.plot(self.x, self.phi_now, label=f'$\phi$({int(t)} s)')
        t = self.t0 + dt
        while t < self.t1:
            phi_new = self.ctcs(self.phi_old, self.phi_now)
            self.d = self.alpha * (self.phi_old + phi_new - 2.0 * self.phi_now)
            #if self.beta == 0:
            #    self.phi_old = self.phi_now
            #    self.phi_now = phi_new + self.d
            #else:
            self.phi_old[:] = self.phi_now + self.beta * self.d
            self.phi_now[:] = phi_new + (1 - self.beta) * self.d
            t += self.dt
            if np.mod(t, self.tp) < self.dt:
                plt.plot(self.x, self.phi_now, label=f'$\phi$({int(t)} s)')
            
        plt.xlabel('x')
        plt.suptitle(r'Leapfrog Scheme $\frac{\partial \phi}{\partial t} = -u \frac{\partial \phi}{\partial x}$', fontsize=13)
        plt.title(rf'$\alpha =$ {self.alpha}, $\beta =$ {self.beta}, $\Delta t =$ {self.dt}, $u =$ {self.u}')
        plt.legend()
        plt.savefig(f'Leapfrog Scheme, alpha = {self.alpha}, beta = {self.beta}.png')
        plt.show()
```

```python
# Define parameters
x0 = 0.0
x1 = 1000.0
t0 = 0.0
t1 = 2000.0
u = 0.475
dx = 0.125
dt = 0.10
tp = 500.0
alpha = 0.05
beta = 0.53

# Create and plot
AdvectionSimulation(x0, x1, t0, t1, u, dx, dt, tp, alpha, beta).plot()
```
